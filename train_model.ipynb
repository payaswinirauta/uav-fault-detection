```python
# Step 1 — Import Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import joblib
import os

# Step 2 — Check Dataset Files
print("Files in dataset folder:", os.listdir("/kaggle/input/uav-multifault-dataset"))

# Step 3 — Load Dataset
df = pd.read_csv("/kaggle/input/uav-multifault-dataset/uav_multifault_dataset.csv")
print("Dataset Loaded ✅")
print(df.head())

# Step 4 — Preprocess Data
# Assuming the target column is 'fault_type'
X = df.drop("fault_type", axis=1)
y = df["fault_type"]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 5 — Train/Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# Step 6 — Train Random Forest Model
model = RandomForestClassifier(n_estimators=200, random_state=42)
model.fit(X_train, y_train)

# Step 7 — Evaluate Model
y_pred = model.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {acc*100:.2f}% ✅")

# Step 8 — Save Model & Scaler
joblib.dump(model, "uav_fault_model.pkl")
joblib.dump(scaler, "scaler.pkl")
print("Model and Scaler saved as .pkl files ✅")

# Step 9 — (Optional) Download .pkl files locally for Fullstack Integration
# In Kaggle, use the 'Output' folder to save files and then download
output_path = "/kaggle/working/"
joblib.dump(model, os.path.join(output_path, "uav_fault_model.pkl"))
joblib.dump(scaler, os.path.join(output_path, "scaler.pkl"))
print("Files saved in /kaggle/working/ folder, ready for download ✅")
```
